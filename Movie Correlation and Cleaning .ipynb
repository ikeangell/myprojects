```python
#Importing libraries
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Reading in data
df = pd.read_csv(r'C:\Users\ikean\Desktop\CSV Files\movies.csv')
```

```python
#Looking at first five rows of dataframe
df.head()
```

```python
#Checking for missing data on a percentage base for each column 
for col in df.columns:
    missing_data_pcts = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_data_pcts))
```

```python
#Dropping rows with missing data 
df.dropna()
```

```python
#Checking for and dropping duplicate movies 
df['name'].drop_duplicates().sort_values(ascending = False)
```

```python
#Checking column data types
df.dtypes
```

```python
#Changing data type for budget, gross, and votes columns 
df['budget'] = pd.to_numeric(df['budget'], errors = 'coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors = 'coerce').fillna(0).astype(int)
df['votes'] = pd.to_numeric(df['votes'], errors = 'coerce').fillna(0).astype(int)
```

```python
df.head()
```

```python
#Splitting the country release from the release date
df[['released','country_release']] = df['released'].str.split(' \(',n=1,expand=True)
```

```python
#Giving the country release it's own column and cleaning it 
df['country_release'] = df['country_release'].str.replace(')','')
```

```python
#Applying a date time format to the release column to match with the correct year in the year column
df['released'] = pd.to_datetime(df['released'],format='mixed')
```

```python
df.head()
```

```python
#Ordering the dataframe by gross revenue in descending order 
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#Scatter plot with budget vs gross revenue 
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget Vs Gross Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
#Regression plot with budget vs gross revenue 
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color':'black'}, line_kws = {'color': 'purple'})
```

```python
#Regression plot with movie score vs gross revenue 
sns.regplot(x = 'score', y = 'gross', data = df, scatter_kws = {'color':'black'}, line_kws = {'color':'purple'})
```

```python
#Regression plot with movie score vs budget
sns.regplot(x = 'score', y = 'budget', data = df, scatter_kws = {'color':'black'}, line_kws = {'color': 'purple'})
```

```python
#Running a correlation matrix to see which columns have strong correlation
correlation_matrix = df.corr(method = 'pearson', numeric_only = True)
```

```python
#Creating a heatmap with correlation matrix 
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
#Turning object columns into numeric columns 
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
df_numeric
```

```python
#Running a correlation matrix on all columns 
df_numeric.corr()
```

```python
#Unstacking above matrix for readability 
correlation_matrixV2 = df_numeric.corr()
correlation_pairs = correlation_matrixV2.unstack()
correlation_pairs
```
